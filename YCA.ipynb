```python
from dataaccess.csv import get_data
from indicators.average import rolling_average, wilder_average
from indicators.true_range import atr
from indicators.rsi import rsi
from trading.signal import cross_from_below
import numpy
from plotly.subplots import make_subplots
import plotly.graph_objects as go
 

df = get_data('COMEX_HG-MT_prices.csv')

short_ma_length = 50
long_ma_length = 200

df['short_ma'] = rolling_average(df.Close.values, short_ma_length)
df['long_ma']  = rolling_average(df.Close.values, long_ma_length)
df['diff'] = df['short_ma'] - df['long_ma']
df['atr'] = atr(df.High.values, df.Low.values, df.Close.values, 20)
df['atrMA'] = wilder_average(df.atr.values, 9, 20)
df['rsi'] = rsi(df.Close.values, 20)
df['rsiMA'] = wilder_average(df.rsi.values, 9, 20)

df['ShortCrossesLong'] = cross_from_below(df.short_ma.values, df.long_ma.values)

print(numpy.nansum(df['ShortCrossesLong'].values))

fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.0,row_heights=[0.8,0.2])

fig.add_trace(go.Scatter(x=df.Date.values, y=df.Close.values, name='Close'), row=1, col=1)
fig.add_trace(go.Scatter(x=df.Date.values, y=df.long_ma.values, name='long_ma'), row=1, col=1)
fig.add_trace(go.Scatter(x=df.Date.values, y=df.short_ma.values, name='short_ma'), row=1, col=1)
fig.add_trace(go.Scatter(x=df.Date.values, y=df['rsi'].values, name='ATR 20'), row=2, col=1 )
fig.add_trace(go.Scatter(x=df.Date.values, y=df['rsiMA'].values, name='ATR MA'), row=2, col=1 )




fig.update_layout(width=1200, height =800)
fig.show()
```
